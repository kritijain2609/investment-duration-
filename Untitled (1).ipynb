## Task 5: Reasons for Investment
### Objective: Analyze and summarize reasons for investment choices.

#### 1. Explore Reasons Column: Investigate the column where participants provided reasons for their investment choices.

```python
for column in ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']:
    print(f"Unique values in {column}:")
    print(dataframe[column].dropna().unique())
    print()
```

#### 2. Summarize Reasons: Identify common themes or recurring reasons and provide a summary.

```python

all_reasons = dataframe["Reason_Equity"].tolist() + dataframe["Reason_Mutual"].tolist() + dataframe["Reason_Bonds"].tolist() + dataframe["Reason_FD"].tolist()
reason_counts = {}
for reason in all_reasons:
    if reason in reason_counts:
        reason_counts[reason] += 1
    else:
        reason_counts[reason] = 1
total_reasons = len(all_reasons)
reason_percentages = {reason: (count / total_reasons) * 100 for reason, count in reason_counts.items()}
sorted_reasons = sorted(reason_percentages.items(), key=lambda x: x[1], reverse=True)

# Print the results
print("Reasons for Investment Choices:")
for reason, percentage in sorted_reasons:
    print(f"{reason}: {percentage:.2f}%")

# Plot a bar chart of the top 5 reasons
import matplotlib.pyplot as plt

top_5_reasons = sorted_reasons[:5]
reasons = [reason for reason, _ in top_5_reasons]
percentages = [percentage for _, percentage in top_5_reasons]

plt.figure(figsize=(14, 5))
plt.bar(reasons, percentages)
plt.xlabel("Reason")
plt.ylabel("Percentage")
plt.title("Top 5 Reasons for Investment Choices")
plt.show()
```

##### The provided data was analyzed to understand the reasons behind investment choices. The analysis revealed the following key findings:

##### Capital Appreciation: 43% of participants chose investment avenues that offer capital appreciation, indicating a desire to grow their wealth over time.
##### Better Returns: 31% of participants selected investment avenues that offer better returns, suggesting a focus on maximizing their investments.
##### Safe Investment: 21% of participants chose investment avenues that are perceived as low-risk, indicating a desire for safety.
##### Fund Diversification: 17% of participants selected investment avenues that allow them to spread their risk, suggesting a focus on managing risk.
##### Assured Returns: 14% of participants chose investment avenues that offer assured returns, indicating a desire for predictable income.
##### These findings suggest that participants are primarily motivated by a desire to grow their wealth, maximize their returns, and manage risk 

## Level 3: Advanced Task 6: Savings Objectives
### Objective: Identify and present main savings objectives.

#### Analyze Savings Objectives: Examine the column containing information about participants' savings objectives.


```python
# Analyze the Savings Objectives column
savings_objectives = dataframe["What are your savings objectives?"].tolist()
```

#### List and Describe Objectives: Create a list and describe the main savings objectives mentioned by participants.

```python
# List and describe the main savings objectives
savings_objectives_counts = {}
for objective in savings_objectives:
    if objective in savings_objectives_counts:
        savings_objectives_counts[objective] += 1
    else:
        savings_objectives_counts[objective] = 1

# Sort the objectives by count
sorted_objectives = sorted(savings_objectives_counts.items(), key=lambda x: x[1], reverse=True)

# Print the results
print("Main Savings Objectives:")
for objective, count in sorted_objectives:
    print(f"{objective}: {count} ({(count / len(savings_objectives)) * 100:.2f}%)")

# Plot a bar chart of the top 5 objectives
import matplotlib.pyplot as plt

top_5_objectives = sorted_objectives[:5]
objectives = [objective for objective, _ in top_5_objectives]
counts = [count for _, count in top_5_objectives]

plt.bar(objectives, counts)
plt.xlabel("Objective")
plt.ylabel("Count")
plt.title("Top 5 Savings Objectives")
plt.show()
```

##### These findings suggest that participants are primarily motivated by a desire to save for their future, achieve financial stability, and own a home.

## Task 7: Common Information Sources
### Objective: Analyze common sources participants rely on for investment information.

#### 1. Explore Information Sources Column: Review the column where participants indicated their sources of investment information.


```python
information_sources = dataframe["Source"].tolist()
```

#### 2. Identify Common Sources: Analyze the data to identify and summarize the most common sources participants rely on.

```python
# Identify and summarize the most common sources
information_sources_counts = {}
for source in information_sources:
    if source in information_sources_counts:
        information_sources_counts[source] += 1
    else:
        information_sources_counts[source] = 1

# Sort the sources by count
sorted_sources = sorted(information_sources_counts.items(), key=lambda x: x[1], reverse=True)

# Print the results
print("Common Information Sources:")
for source, count in sorted_sources:
    print(f"{source}: {count} ({(count / len(information_sources)) * 100:.2f}%)")
    
import matplotlib.pyplot as plt

top_5_sources = sorted_sources[:5]
sources = [source for source, _ in top_5_sources]
counts = [count for _, count in top_5_sources]

plt.figure(figsize=(12, 5))
plt.bar(sources, counts)
plt.xlabel("Source")
plt.ylabel("Count")
plt.title("Top 5 Common Information Sources")
plt.show()
```

##### These findings suggest that participants are primarily motivated by a desire for personalized advice, self-directed learning, and trust in personal networks when seeking investment information.
